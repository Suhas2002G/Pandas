### Pandas Series
A series is a one-dimensional labeled array capable of holding any datatype. 
The axis labels are collectively called the index

##### Creating a Series
There are multiple ways to create a series

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a','b','c']
my_list = [10,20,30]
arr = np.array([10,20,30])
d = {10:100, 20:200, 30:300}
```

```python
# when we pass only datapoints then automatically index start from 0, these index nothing but labels
pd.Series(data=my_list)
```

```python
# we can also set custom labels by providing labels parameter
pd.Series(data=my_list, index=labels)
```

```python
# we can also create a series using numpy array
pd.Series(arr)
```

```python
# create a series using dictionary
pd.Series(d)  # here keys are labels and values and series element
```
